```python
class TarjetaCredito:
    def __init_(self, banco, limite):
        self.banco = banco
        self.limite = limite

    def mostrar_limite(self):
        print(f"Tarjeta de {self.banco} con limite de ${self.limite}")

class UsuarioConTarjeta:
    def __init__(self, nombre, tarjeta):
        self.nombre = nombre

    def mostrar_tarjeta(self):
      print(f"usuario: {self.nombre}")
      self.tarjeta.mostrar_limite()

mi_tarjeta = TarjetaCredito("banco demo", 100000)
usuario = UsuarioConTarjeta("Ana", mi_tarjeta)
usuario.mostrar_tarjeta()
```
